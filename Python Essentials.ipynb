```python
from collections import defaultdict, Counter

#Sorting

x = [4,1,2,3]
y = sorted(x) #this leaves x unchanged

x.sort() # this changes x

#Sort list from largest to smallest:

x1 = sorted([-4,1,-2,3], key = abs, reverse=True)
print(x1)
```

```python
#Dictionaries

#defaultdict is like a regular dictionary except that when you try to look up 
#a key it doesn't contain, it first adds a value for it using a zero-argument function you provided when you created it

dd_list = defaultdict(list)
dd_list[2].append(1)

print(dd_list.get(2))

dd_pair = defaultdict(lambda: [0,0])

dd_pair[3][0] = 12
dd_pair[2][1] =6

print(dd_pair)

c = Counter(dd_pair)

print(c)
```

```python


#Sets - Data structures which represent a collection of distinct elements

s = set()
s.add(1)
s.add(2)
print(len(s))
print(2 in s)
print(3 in s)

#Sets are used because in is a very fast operation on sets - fast and easy membership test

#Sets are also used to find distinct elements

item_list = [1,2,3,1]
item_set = set(item_list)

print(len(item_list))
print(f"Set has length {len(item_set)} whereas list has length {len(item_list)}")
print(item_set)
```

```python
#List comprehensions

even_numbers = [x for x in range(5) if x % 2 == 0]
squares = [x * x for x in range(5)]
even_squares = [x *x for x in even_numbers]

print(even_squares)

pairs = [(x,y) for x in range(4)
         for y in range(4)]

print(pairs)
```

```python
import random

# Randomness is deterministic, decided by internal state which can be set with random.seed(n)
# random.seed(3)
four_uniform_randoms = [random.random() for _ in range(4)]
print(four_uniform_randoms)

print(random.randrange(10))
print(random.randrange(3,6))

up_to_ten = [x for x in range(10)]
random.shuffle(up_to_ten)
print(up_to_ten)

print(random.choice(up_to_ten))

#choosing without replacement

print(random.sample(up_to_ten, 6))

#choosing with replacement

four_with_replacement = [random.choice(range(10)) for _ in range(4)]
print(four_with_replacement)
```

```python
#zip and Argument unpacking

list1 = ['a','b','c']
list2 = [1,2,3]

x = zip(list1, list2)
print(tuple(x))


# We can also unzip. * performs argument unpacking which uses the elements of pairs as individual arguments to zip

pairs = [('a', 1), ('b', 2), ('c', 3)]

letters, numbers = zip(*pairs)
print(letters)
print(list(numbers))
```

```python
#Defining functions of functions

#args is a tuple of unnamed arguments, kwargs is a dict of its named arguments

# def magic(*args, **kwargs):

def doubler(f):
    def g(*args,**kwargs):
        return 2 * f(*args, **kwargs)
    return g

def f2(x, y):
    return x + y

g = doubler(f2)
print(g(1,4))
```

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]

total_error = [x + y for x, y in zip(variance,bias_squared)]

xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'gx-', label="variance")
plt.plot(xs, bias_squared, 'r-', label = "bias_squared")
plt.plot(xs, total_error, 'b:', label = "total_error")
plt.title("The Bias-Variance Tradeoff")
plt.legend(loc=9)j
```

```python
from functools import reduce

def vector_add(v,w):
    return [v_i + w_i for v_i, w_i in zip(v,w)]

vector_add([1,2],[3,3])

#reduce(fun,seq) is used to apply a particular function passed in its argumet to all of the list elements:

def vector_sum(vectors):
    return reduce(vector_add, vectors)

vector_sum(([1,2],[3,4],[5,6]))
```
